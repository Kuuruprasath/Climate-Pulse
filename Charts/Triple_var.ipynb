```python
import plotly.express as px
import sys
import os
import pandas as pd
sys.path.append('c:/Users/kuuru/3164/Climate-Pulse/Conversions')
import convertt
import plotly.graph_objects as go

```

```python
dataset = convertt.suburb_to_var(["Alpine"])

```

```python
def lineChart(dataset,variable,startDate,endDate,period):


    dataset["datetime"] = pd.to_datetime(dataset["datetime"])
    daterange = pd.date_range(startDate, endDate, freq=period)

    filtered_dataset = dataset[dataset["datetime"].isin(daterange)]

    fig = px.line(filtered_dataset, y=variable,x = "datetime")
    fig.update_layout(plot_bgcolor='white')
    fig.show()

startDate = pd.Timestamp("2015-01-01", tz='UTC')
endDate = pd.Timestamp("2024-07-31", tz='UTC')
variable = ["temperaturemean","rainsum","relativehumiditymean"]
period = "M"
print(lineChart(dataset,variable,startDate,endDate,period))
```

```python
def violinChart(dataset):

    fig = go.Figure()

    fig.add_trace(go.Violin(y=dataset['temperaturemean'], name='Temperature', box_visible=True, meanline_visible=True))
    fig.add_trace(go.Violin(y=dataset['relativehumiditymean'], name='Humidity', box_visible=True, meanline_visible=True))
    fig.add_trace(go.Violin(y= dataset['rainsum'], name='Rainfall', box_visible=True, meanline_visible=True))

    # Update layout
    fig.update_layout(title='Violin Plot of Temperature, Humidity, and Rainfall',
                    yaxis_title='Values',
                    xaxis_title='Weather Parameters')

    # Show the plot
    fig.show()

print(violinChart(dataset))
```

```python
def barChart(dataset,variable,startDate,endDate,period):


    dataset["datetime"] = pd.to_datetime(dataset["datetime"])
    daterange = pd.date_range(startDate, endDate, freq=period)

    filtered_dataset = dataset[dataset["datetime"].isin(daterange)]

    fig = px.bar(filtered_dataset, y=variable,x = "datetime",barmode="group")
    fig.update_layout(plot_bgcolor='white')
    fig.show()

startDate = pd.Timestamp("2000-01-01", tz='UTC')
endDate = pd.Timestamp("2024-07-31", tz='UTC')
variable = ["temperaturemean","relativehumiditymean","rainsum"]
period = "Y"
print(barChart(dataset,variable,startDate,endDate,period))
```

```python
def areachart(dataset,variable,startDate,endDate,period):


    dataset["datetime"] = pd.to_datetime(dataset["datetime"])
    daterange = pd.date_range(startDate, endDate, freq=period)

    filtered_dataset = dataset[dataset["datetime"].isin(daterange)]

    fig = px.area(filtered_dataset,x = "datetime",y = variable)
    fig.update_layout(plot_bgcolor='white')

    fig.show()

startDate = pd.Timestamp("2000-01-01", tz='UTC')
endDate = pd.Timestamp("2024-07-31", tz='UTC')
variable = ["temperaturemean","relativehumiditymean","rainsum"]
period = "Y"
print(areachart(dataset,variable,startDate,endDate,period))
```
