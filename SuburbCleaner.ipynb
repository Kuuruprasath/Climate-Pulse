```python
import pandas as pd
import numpy as np
df = pd.read_csv("Datasets/georef-australia-state-suburb.csv", delimiter=';')
```

```python
df.head()
```

```python
# 1 missing value from geo point
df.info()
```

```python
df[df["Geo Point"].isna()]
```

```python
# Add manually 
# https://geohack.toolforge.org/geohack.php?pagename=Corinna,_Tasmania&params=41.650798_S_145.078356_E_type:city_region:AU-TAS
df.loc[df["Geo Point"].isna(), ["Geo Point"]] = "-41.650798, 145.078356"
```

```python
cord = df["Geo Point"].str.split(',').to_list()
cord = np.array(cord, dtype=float)
```

```python
df["Latitude"] = cord[:, 0]
df["Longitude"] = cord[:, 1]
```

```python
df.head()
```

```python
df_out = df.loc[:, ["Official Name Suburb", "Official Name State", "Official Code Local Government Area",
            "Official Code State", "Latitude", "Longitude", "Geo Shape"]]
df_out.columns = df_out.columns.str.replace(' ', '') 
```

```python
df_out.to_csv("Datasets/Suburb.csv")
```
