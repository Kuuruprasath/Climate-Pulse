```python
import pandas as pd
import numpy as np
df = pd.read_csv("../Datasets/georef-australia-state-suburb.csv", delimiter=';')
```

```python
df.head()
```

```python
# 1 missing value from geo point
df.info()
```

```python
# Remove whitespace from the columns name
df.columns = df.columns.str.replace(' ', '')
```

```python
df[df["GeoPoint"].isna()]
```

```python
# One row with a missing value. We can drop this
df = df.dropna()
```

```python
df.groupby("OfficialNameState").count()
```

```python
cord = df["Geo Point"].str.split(',').to_list()
cord = np.array(cord, dtype=float)
```

```python
df["Latitude"] = cord[:, 0]
df["Longitude"] = cord[:, 1]
```

```python
df.head()
```

```python
df_out = df.loc[:, ["Official Name Suburb", "Official Name State", "Official Code Local Government Area",
            "Official Code State", "Latitude", "Longitude", "Geo Shape"]]
df_out.columns = df_out.columns.str.replace(' ', '') 
```

```python
df_out.to_csv("../Datasets/Suburb.csv", index=False)
```
